```python
alice_img = Image.open('alice_mask.png')
alice_mask = np.array(alice_img)

alice_wc = wordcloud.WordCloud(
    background_color="white",
    max_words=3000,
    mask=alice_mask,
    stopwords=en_stopwords
)
alice_wc.generate(alice)

plt.figure(figsize=(10, 15))
plt.imshow(alice_wc, cmap=plt.cm.gray)
plt.axis("off")
plt.show()
```

```python
df_canada = pd.read_excel(
    "https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
    sheet_name = 'Canada by Citizenship',
    skiprows = range(20),
    skipfooter = 2
)
```

```python
df_canada['total'] = df_canada.sum(axis=1)
df_canada.head()
```

```python
max_words = 100
text = ""
for country in df_canada.OdName.values:
    if len(country.split(" ")) == 1:
        repeat_num_times = int(
            df_canada.loc[
                df_canada['OdName']==country,
                "total"
            ] / float(df_canada['total'].sum()) * max_words
        )
        text = text + ((country + ' ') * repeat_num_times)
```

```python
int(round(df_canada.loc[df_canada.OdName == "China", "total"]/df_canada['total'].sum() * 100))
```

```python
text
```

```python
"United Kingdom".split()
```

```python
immigrant_wc = wordcloud.WordCloud(background_color='white').generate(text)
plt.figure(figsize=(14,9))
plt.imshow(immigrant_wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Seaborn

```python
years = list(range(1980, 2014))
df_total = pd.DataFrame(df_canada[years].sum(axis=0))
df_total = df_total.reset_index()
df_total.columns = ['year', 'total']
df_total.head()
```

```python
# plt.style.use("ggplot")
sns.regplot(x="year", y="total", data=df_total, color='green', marker='+', scatter_kws={"s": 100, "linewidth":5})
plt.title("Total Immigrants per Year (1980 - 2014)")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.regplot(x="year", y="total", data=df_total, color='green', marker='.')
plt.title("Total Immigrants per Year (1980 - 2014)")
plt.show()
```

```python
sf_incidents = pd.read_csv("http://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv")
```

```python
sf_incidents.head()
```

```python
subset_sf_incidents = sf_incidents.loc[:50]
```

```python
sf_lat = 37.77
sf_long = -122.42

sf_map = folium.Map(location=(sf_lat,sf_long), zoom_start=12)
sf_map
```

```python
#fungsi zip
a = [1, 2, 3]
b = [10, 50, 60]

for i, j in zip(a,b):
    print(i, j)
```

```python
incidents = folium.map.FeatureGroup()

for lat, long in zip(subset_sf_incidents.Y, subset_sf_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            location=(lat, long),
            radius=5,
            color="yelow",
            fill=True,
            fill_color="blue",
            fill_opacity=0.6
        )
    )
sf_map.add_child(incidents)
sf_map
```

```python
incidents = folium.map.FeatureGroup()

for lat, long in zip(subset_sf_incidents.Y, subset_sf_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            location=(lat, long),
            radius=5,
            color="yelow",
            fill=True,
            fill_color="blue",
            fill_opacity=0.6
        )
    )
labels = sf_incidents.Category.tolist()
for lat, long, label in zip(subset_sf_incidents.Y, subset_sf_incidents.X, labels):
    folium.Marker((lat,long), popup=label).add_to(sf_map)
    
sf_map.add_child(incidents)
sf_map
```
