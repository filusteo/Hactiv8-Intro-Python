```python
import numpy as np
import pandas as pd
```

```python
df_canada = pd.read_excel(
    "https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
    sheet_name = 'Canada by Citizenship',
    skiprows = range(20),
    skipfooter = 2
)
```

```python
df_canada.shape
```

```python
type(df_canada)
```

```python
df_canada.columns.values
```

```python
df_canada.drop(
    columns=[
        "AREA", "REG", "DEV", "Type", "Coverage"
    ], axis = 1, inplace = True
)
```

```python
df_canada.head()
```

```python
df_canada.rename(
    columns = {
        'OdName':'Country',
        'AreaName':'Continent',
        'RegName':'Region'
    }, inplace=True
)
```

```python
df_canada['Total'] = df_canada.sum(axis=1)
df_canada.head()
```

```python
df_canada.isnull().sum()
```

```python
plt.style.use("fivethirtyeight")
# plt.style.use("ggplot")
```

```python
years = list(range(1980,2004))

haiti = df_canada.loc[df_canada["Country"] == 'Haiti', years].transpose()
haiti
```

```python
haiti.plot()
plt.ylabel("Number of Imigration from Haiti")
plt.xlabel("Years")
plt.title("Immigration from Haiti over Years")
# plt.text(1998, 6000, "2010 Earthaquake")

plt.show()
```

```python
cn_in = df_canada.loc[
    df_canada["Country"].isin(["China", "India"]),
    years
].transpose()
cn_in.rename(
    columns={
        36: "China",
        79: "India"
    }, inplace=True
)
cn_in
```

```python
cn_in.plot()
```

```python
top5_countries = df_canada.sort_values(
    by="Total", ascending=False, axis=0
).head()
top5_countries
```

```python
top5_countries = top5_countries.set_index("Country")
top5_countries = top5_countries[years].transpose()
```

```python
top5_countries.plot(figsize=(15,9))
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")

plt.show()
```

```python
top5_countries.plot(
    kind="area",
    alpha=.25,
    stacked=True,
    figsize=(10,10)
)
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")

plt.show()
```

```python
#object oriented
ax = top5_countries.plot(
    kind="area",
    alpha=.25,
    stacked=True,
    figsize=(10,10)
)
ax.set_title("Top 5 Immigrant Origin over Years in Canada")
ax.set_ylabel("Num of Immigrant")
ax.set_xlabel("Years")

plt.show()
```

```python
least5_countries = df_canada.sort_values(
    by="Total", ascending=True, axis=0
).head(5)
least5_countries = least5_countries.set_index("Country")
least5_countries = least5_countries[years].transpose()
least5_countries.plot(
    kind="area",
    alpha=.25,
    stacked=False,
    figsize=(10,10)
)
plt.title("Least 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")

plt.show()
```
