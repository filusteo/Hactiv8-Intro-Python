## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

Bagaimana perkembangan atau banyaknya jumlah kriminal dari tahun ke tahun berdasarkan major criminal category?

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
major_category_unique = df.major_category.value_counts()
major_value_year = df.loc[:, ["major_category", "year", "value"]]
sum_criminal = major_value_year.groupby(["major_category", "year"], as_index=False, axis=0)["value"].sum()
df_sum_criminal = sum_criminal.pivot(index='major_category', columns='year', values='value')
df_sum_criminal = df_sum_criminal.transpose()

df_sum_criminal.plot(
    kind="area",
    alpha= .20,
    stacked=False,
    figsize=(15,9)
)
# Graded-Funtion End
plt.title('Area Plot of Major Category Over Years') # add a title to the area plot
plt.ylabel('Num of Crimes') # add y-label
plt.xlabel('Year') # add x-label

plt.show()
```

**Insight:**

1. Berdasarkan area plot diatas dapat dilihat bahwa major kategori kriminal yang paling banyak terjadi adalah 'Theft and Handling'
2. Puncak dari banyaknya 'Theft and Handling' terjadi pada tahun 2012


## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

Bagaimana distribusi frekuensi kejadian kriminal berdasarkan wilayah dari tahun 2008 sampai 2016?

```python
# Write your function below

# Graded-Funtion Begin (~2 Lines)
df_crime_by_borough = df.groupby("borough", axis=0)["value"].sum()

df_crime_by_borough.plot(
    kind="hist",
    alpha=.6,
    figsize=(15,9)
)
# Graded-Funtion End

plt.title('Histogram of Crimes by Borough') # add a title to the histogram
plt.ylabel('Number of Borough') # add y-label
plt.xlabel('Number of Crime') # add x-label

plt.show()
```

**Insight:**

Dari distribusi diatas dapat dilihat bahwa :
1. Hanya satu wilayah yang kemunculan tindak kriminal lebih dari 400000 kali.
2. Tidak ada wilayah yang memiliki tindak kriminal 300000 kali.
3. Distribusi yang terjadi dari tahun 2008 sampai 2016 dapat dikatakan berdistribusi normal karena tidak skew ke kiri maupun ke kanan. Namun memang terdapat satu wilayah dengan tindak kriminal yang terjadi lebih tinggi dari yang lainnya.
4. Pada case ini tidak ada satupun wilayah yang tidak memiliki kejadian tindakan kriminal.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

Bagaimana jumlah tindak kriminal yang terjadi pada wilayah paling banyak terjadi tindak kriminal yaitu Westminster untuk setiap kategorinya ?

```python
df_crime_by_borough
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_westminster = df[df['borough'] == 'Westminster']
minor_category = df_westminster.groupby("minor_category", axis=0)["value"].sum()
major_category = df_westminster.groupby("major_category", axis=0)["value"].sum()


fig = plt.figure(figsize=(20,6))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(1, 2, 2)

minor_category.plot(kind='bar', ax=ax1)
ax1.set_title("Minor Category of Crime in Westminster")
ax1.set_xlabel("Minor Category")
ax1.set_ylabel("Number of Crimes")

major_category.plot(kind="bar", ax=ax2)
ax2.set_title("Major Category of Crime in Westminster")
ax2.set_xlabel("Major Category")
ax2.set_ylabel("Number of Crimes")


# Graded-Funtion End

# plt.xlabel('Year') # add to x-label to the plot
# plt.ylabel('Number of Crimes') # add y-label to the plot
# plt.title('Theft and Handling Crimes over Years') # add title to the plot

plt.show()
```

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Bagaimana persentase jumlah kejadian tindak kriminal untuk kategori major ?

```python
df_major = df.groupby('major_category', axis=0).sum()
df_major
```

```python
# Write your function below

# ratio for each continent with which to offset each wedge.
explode_list = [0, 0, 0, 0, 0, 0, 0, 0, 0]

# Graded-Funtion Begin (~8 Lines)
df_major['value'].plot(kind='pie',
                    figsize=(15, 6),
                    autopct='%1.1f%%',
                    startangle=60,
                    shadow=False,
                    labels=None,         # turn off labels on pie chart
                      # the ratio between the center of each pie slice and the start of the text generated by autopct
                    pctdistance=1.2,
                    fontsize=9,
#                       subplots=True,
                    explode=explode_list  # 'explode'
                      )
# Graded-Funtion End

# scale the title up by 12% to match pctdistance
plt.title('Persentage of Major Category Crime', y=1.12)

plt.axis('equal')

# add legend
plt.legend(
    labels=df_major.index,
    loc="upper left",
    fontsize='x-small'
)
plt.show()
```

**Insight:**

Dari pie chart diatas dapat dilihat bahwa:
1. Persentase masing masing tindak kejahatan tidak sama, yaitu dengan tindak kriminal yang paling banyak sebesar 41.3% sedangkan untuk persentase tindak kriminal paling sedikit 0.1%
2. Tiga tindak kriminal yang paling banyak terjadi adalah 'Theft and Handling', 'Burglary', dan 'Violence Against the Person'
3. Tindak kriminal yang paling sedikit berada diangka kurang dari 1% sehingga perbedaan dengan tindak kriminal dengan yang lainnya tidak terlihat

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

Bagaimana persebaran jumlah tindakan kriminal pada kategori major dari tahun 2008 sampai 2016?

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_sum_criminal.plot(
            kind="box",
            figsize=(20, 6)
        )

# Graded-Funtion End
plt.title('Box Plot of Major Category Crime over Years')
plt.ylabel('value')

plt.show()
```

**Insight:**

Dari boxplot di atas dapat dilihat bahwa :
1. Untuk tindak kriminal 'Burglary', 'Criminal Damage', 'Drugs', 'Other Notifiable Offences', 'Robery', 'Theft and Handling', dan 'Violance Against the Person' tidak terdapat outlier pada jumlah dari tindak kriminal
2. Masing-masing tindakan kriminal memiliki jumlah max dan minimum yang berbeda beda
3. Untuk tindak kriminal 'Fraud or Forgery', dan 'Sexual Offences' terlihat indikasi outlier namun apabila dilihat lagi nilai tersebut merupakan jumlah dari tindak kriminal satu-satunya. Jadi tindak kriminal tersebut hanya terjadi satu kali pada salah satu tahun dari 2008 sampai 2016.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Bagaimana persebaran jumlah tindakan kejahatan dari tahun 2008 sampai 2016?

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines
crime_by_year = df.groupby("year", as_index=False, axis=0)["value"].sum()
crime_by_year.plot(kind='scatter', x='year', y='value', figsize=(10, 6), color='darkblue')
# Graded-Funtion End

plt.title('Scatter Plot of Crimes over Year')
plt.xlabel('Year')
plt.ylabel('Num of Crime')

plt.show()
```

**Insight:**

Dari boxplot di atas dapat dilihat bahwa :
1. Tahun dengan jumlah tindak kriminal paling banyak adalah tahun 2008, 2012 dan 2016
2. Tahun dengan jumlah tindak kriminal paling sedikit adalah tahun 2014, dan 2013

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)
list_minor = list(df['minor_category'])
def listToString(s):  
    str1 = " "    
    return (str1.join(s))   
minor_string= listToString(list_minor)
# generate the word cloud
your_wordcloud.generate(minor_string)
your_wordcloud
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
plt.imshow(your_wordcloud, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```
